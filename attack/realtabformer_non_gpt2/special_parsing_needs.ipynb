```python
import re

def extract_booking_info(text):
    pattern = re.compile(
        r"A guest at ###(?P<guest_email>[^#]+)### booked a ###(?P<room_type>[^#]+)### room at ###(?P<hotel_id>[^#]+)### from ###(?P<checkin_date>[^#]+)### to ###(?P<checkout_date>[^#]+)###, for ###(?P<room_rate>[^#]+)###, with an amenities fee of ###(?P<amenities_fee>[^#]+)### and a billing address of ###(?P<billing_address>[^#]+)###\. They used a credit card with the number ###(?P<credit_card_number>[^#]+)### and have ###(?P<has_rewards>[^#]+)### rewards membership."
    )
    
    matches = pattern.findall(text)
    extracted_data = []
    for match in matches:
        extracted_data.append({
            "guest_email": match[0],
            "room_type": match[1],
            "hotel_id": match[2],
            "checkin_date": match[3],
            "checkout_date": match[4],
            "room_rate": match[5],
            "amenities_fee": match[6],
            "billing_address": match[7],
            "credit_card_number": match[8],
            "has_rewards": match[9],
        })
    
    return extracted_data

# Example usage
text = df['Answer'][0].split('###)')[-1]

booking_info = extract_booking_info(text)[0]
booking_info
output = []

for i in range(len(df)):
    text = df['Answer'][i].split('###)')[-1]
    booking_info = extract_booking_info(text)
    if len(booking_info) > 0:
        output.append(extract_booking_info(text)[0])

pd.DataFrame(output)
```
